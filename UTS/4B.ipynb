# Soal 4B - Forward & Backward Pass ANN

**Data yang digunakan (4B):**

| Sample | I1 | I2 | y  |
|--------|----|----|----|
| 1      | -4 |  3 | -1 |
| 2      |  1 |  0 |  5 |

**Konfigurasi jaringan:**
- Arsitektur: 2 input -> Hidden Layer 1 (3 neuron) -> Hidden Layer 2 (2 neuron) -> 1 output
- Bias = 1 untuk semua neuron
- Aktivasi hidden layer: ReLU
- Aktivasi output: Linear (tidak ada aktivasi)
- Loss function: MSE
- Learning rate: 0.01
- Metode: Batch Gradient Descent (kedua sampel diproses sekaligus)


## Tabel Bobot Awal

| Layer        | Bobot  | Nilai | Koneksi       |
|--------------|--------|-------|---------------|
| Input -> H1  | W1_h1  |  2    | I1 -> H1_1    |
| Input -> H1  | W2_h1  | -1    | I2 -> H1_1    |
| Input -> H1  | W3_h1  |  1    | I1 -> H1_2    |
| Input -> H1  | W4_h1  |  1    | I2 -> H1_2    |
| Input -> H1  | W5_h1  | -2    | I1 -> H1_3    |
| Input -> H1  | W6_h1  | -1    | I2 -> H1_3    |
| H1 -> H2     | W1_h2  |  1    | H1_1 -> H2_1  |
| H1 -> H2     | W2_h2  |  3    | H1_2 -> H2_1  |
| H1 -> H2     | W3_h2  | -2    | H1_3 -> H2_1  |
| H1 -> H2     | W4_h2  |  1    | H1_1 -> H2_2  |
| H1 -> H2     | W5_h2  |  2    | H1_2 -> H2_2  |
| H1 -> H2     | W6_h2  | -2    | H1_3 -> H2_2  |
| H2 -> Output | W1_o   | -2    | H2_1 -> O     |
| H2 -> Output | W2_o   |  3    | H2_2 -> O     |

## A. Simulasi Manual - 1 Forward Pass + 1 Backward Pass


### A.1 Forward Pass

---

#### Sample 1 (I1 = -4, I2 = 3, y = -1)

**Hidden Layer 1 - aktivasi ReLU**

z_H1_1 = W1_h1 * I1 + W2_h1 * I2 + b
       = (2)(-4) + (-1)(3) + 1
       = -8 - 3 + 1 = -10

a_H1_1 = ReLU(-10) = 0

z_H1_2 = W3_h1 * I1 + W4_h1 * I2 + b
       = (1)(-4) + (1)(3) + 1
       = -4 + 3 + 1 = 0

a_H1_2 = ReLU(0) = 0

z_H1_3 = W5_h1 * I1 + W6_h1 * I2 + b
       = (-2)(-4) + (-1)(3) + 1
       = 8 - 3 + 1 = 6

a_H1_3 = ReLU(6) = 6

**Hidden Layer 2 - aktivasi ReLU**

z_H2_1 = W1_h2 * a_H1_1 + W2_h2 * a_H1_2 + W3_h2 * a_H1_3 + b
       = (1)(0) + (3)(0) + (-2)(6) + 1
       = 0 + 0 - 12 + 1 = -11

a_H2_1 = ReLU(-11) = 0

z_H2_2 = W4_h2 * a_H1_1 + W5_h2 * a_H1_2 + W6_h2 * a_H1_3 + b
       = (1)(0) + (2)(0) + (-2)(6) + 1
       = 0 + 0 - 12 + 1 = -11
       
a_H2_2 = ReLU(-11) = 0

**Output Layer - aktivasi Linear**

z_O = W1_o * a_H2_1 + W2_o * a_H2_2 + b
    = (-2)(0) + (3)(0) + 1
    = 1

y_hat_1 = 1

Loss_1 = (y - y_hat)^2 = (-1 - 1)^2 = 4

---

#### Sample 2 (I1 = 1, I2 = 0, y = 5)

**Hidden Layer 1 - aktivasi ReLU**

z_H1_1 = (2)(1) + (-1)(0) + 1 = 3   ->   a_H1_1 = ReLU(3) = 3
z_H1_2 = (1)(1) + (1)(0) + 1  = 2   ->   a_H1_2 = ReLU(2) = 2
z_H1_3 = (-2)(1) + (-1)(0) + 1 = -1  ->   a_H1_3 = ReLU(-1) = 0

**Hidden Layer 2 - aktivasi ReLU**

z_H2_1 = (1)(3) + (3)(2) + (-2)(0) + 1 = 3 + 6 + 0 + 1 = 10
a_H2_1 = ReLU(10) = 10

z_H2_2 = (1)(3) + (2)(2) + (-2)(0) + 1 = 3 + 4 + 0 + 1 = 8
a_H2_2 = ReLU(8) = 8

**Output Layer - aktivasi Linear**

z_O = (-2)(10) + (3)(8) + 1 = -20 + 24 + 1 = 5

y_hat_2 = 5

Loss_2 = (5 - 5)^2 = 0

---

#### Total MSE Loss (n = 2)

MSE = (1/n) * (Loss_1 + Loss_2)
    = (1/2) * (4 + 0)
    = 2


### A.2 Backward Pass

Backward pass menggunakan chain rule untuk menghitung gradien setiap bobot terhadap loss.

**Rumus gradien di output layer:**

delta_O = (2/n) * (y_hat - y)

Karena aktivasi output adalah linear, turunannya = 1, jadi tidak ada faktor tambahan.

| Sample | y_hat | y  | delta_O                    |
|--------|-------|----|----------------------------|
| 1      | 1     | -1 | (2/2) * (1 - (-1)) = 2     |
| 2      | 5     |  5 | (2/2) * (5 - 5)    = 0     |

---

**Gradien bobot output layer:**

dL/dW1_o = (1/n) * sum( delta_O * a_H2_1 )
         = (1/2) * [ (2)(0) + (0)(10) ]
         = 0

dL/dW2_o = (1/n) * sum( delta_O * a_H2_2 )
         = (1/2) * [ (2)(0) + (0)(8) ]
         = 0

dL/db_O  = (1/n) * sum( delta_O )
         = (1/2) * (2 + 0)
         = 1

---

**Propagasi error ke Hidden Layer 2:**

Turunan ReLU: ReLU'(z) = 1 jika z > 0, dan 0 jika z <= 0

delta_H2_j = delta_O * W_j_o * ReLU'(z_H2_j)

Sample 1 (delta_O = 2):

  delta_H2_1 = 2 * W1_o * ReLU'(z_H2_1)
             = 2 * (-2) * ReLU'(-11)
             = 2 * (-2) * 0   <- z = -11 < 0, jadi ReLU' = 0
             = 0

  delta_H2_2 = 2 * W2_o * ReLU'(z_H2_2)
             = 2 * (3) * ReLU'(-11)
             = 2 * 3 * 0
             = 0

> Catatan: H2_1 dan H2_2 pada sample 1 punya z = -11 (negatif), sehingga ReLU-nya 0.
> Ini membuat gradien tidak bisa mengalir ke belakang meskipun ada error di output.
> Fenomena ini disebut "Dying ReLU".

Sample 2 (delta_O = 0):

  delta_H2_1 = 0 * (-2) * ReLU'(10) = 0
  delta_H2_2 = 0 * (3) * ReLU'(8)   = 0

> Sample 2 tidak punya error (y_hat = y = 5), jadi delta_O = 0 dari awal.

Karena semua delta_H2 = 0, maka seluruh gradien untuk W_h2, b_H2, W_h1, b_H1 juga = 0.

---

**Ringkasan gradien semua parameter:**

| Parameter     | Gradien | Alasan                                          |
|---------------|---------|-------------------------------------------------|
| dL/dW1_o      | 0       | a_H2_1 = 0 di sample 1, delta_O = 0 di sample 2|
| dL/dW2_o      | 0       | a_H2_2 = 0 di sample 1, delta_O = 0 di sample 2|
| dL/db_O       | 1       | Satu-satunya gradien yang tidak nol             |
| Semua W_h2    | 0       | Dying ReLU di H2 memblokir aliran gradien       |
| Semua W_h1    | 0       | Sama - gradien tidak sampai ke sini             |
| Semua bias H1, H2 | 0   | Sama                                            |

---

**Update bobot (rumus: W_baru = W_lama - lr * gradien):**

| Parameter  | Nilai Lama | Gradien | Update          | Nilai Baru |
|------------|------------|---------|-----------------|------------|
| W1_o       | -2         | 0       | -0.01 * 0 = 0   | -2         |
| W2_o       |  3         | 0       | -0.01 * 0 = 0   |  3         |
| b_O        |  1         | 1       | -0.01 * 1 = -0.01 | 0.99     |
| Semua W_h2 | (masing2)  | 0       | 0               | tidak berubah |
| Semua W_h1 | (masing2)  | 0       | 0               | tidak berubah |

Hanya bias output (b_O) yang berubah setelah 1 step training ini: dari 1.0 menjadi 0.99.

## B. Verifikasi Python (Tanpa Deep Learning Framework)

Semua perhitungan di bagian A diulang menggunakan Python + NumPy saja,
lalu hasilnya dibandingkan satu per satu untuk memastikan tidak ada kesalahan hitung.


```python
import numpy as np

# bobot input -> hidden layer 1, shape (3, 2)
# baris = neuron H1, kolom = input (I1, I2)
W_h1 = np.array([
    [ 2.0, -1.0],   # H1_1
    [ 1.0,  1.0],   # H1_2
    [-2.0, -1.0],   # H1_3
])

# bobot hidden layer 1 -> hidden layer 2, shape (2, 3)
W_h2 = np.array([
    [1.0,  3.0, -2.0],   # H2_1
    [1.0,  2.0, -2.0],   # H2_2
])

# bobot hidden layer 2 -> output, shape (1, 2)
W_o = np.array([[-2.0, 3.0]])

# bias semua = 1
b_h1 = np.ones((3, 1))
b_h2 = np.ones((2, 1))
b_o  = np.ones((1, 1))

# data input X shape (2, 2): baris = fitur (I1, I2), kolom = sampel
X = np.array([[-4.0, 1.0],
              [ 3.0, 0.0]])

# target
Y = np.array([[-1.0, 5.0]])

lr = 0.01
n  = X.shape[1]  # 2 sampel

print("=== DATA ===")
print(f"X:\n{X}")
print(f"Y: {Y}")
print(f"n (jumlah sampel): {n}")
```

### B.1 Forward Pass


```python
def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)

# hidden layer 1
Z_h1 = W_h1 @ X + b_h1
A_h1 = relu(Z_h1)

# hidden layer 2
Z_h2 = W_h2 @ A_h1 + b_h2
A_h2 = relu(Z_h2)

# output layer (linear)
Z_o   = W_o @ A_h2 + b_o
Y_hat = Z_o

print("=" * 50)
print("HASIL FORWARD PASS")
print("=" * 50)

print("\n-- Hidden Layer 1 --")
print(f"Z_h1 (sebelum ReLU):\n{Z_h1}")
print(f"A_h1 (setelah ReLU):\n{A_h1}")

print("\n-- Hidden Layer 2 --")
print(f"Z_h2 (sebelum ReLU):\n{Z_h2}")
print(f"A_h2 (setelah ReLU):\n{A_h2}")

print("\n-- Output Layer --")
print(f"Y_hat (prediksi): {Y_hat}")
print(f"Y     (aktual)  : {Y}")

mse = np.mean((Y - Y_hat) ** 2)
print(f"\nMSE Loss = {mse}")
```

### B.2 Backward Pass


```python
# delta output layer
delta_o = (2 / n) * (Y_hat - Y)

# gradien bobot output
dW_o = (1 / n) * (delta_o @ A_h2.T)
db_o = (1 / n) * np.sum(delta_o, axis=1, keepdims=True)

# propagasi ke H2
delta_h2 = (W_o.T @ delta_o) * relu_derivative(Z_h2)

# gradien bobot H2
dW_h2 = (1 / n) * (delta_h2 @ A_h1.T)
db_h2 = (1 / n) * np.sum(delta_h2, axis=1, keepdims=True)

# propagasi ke H1
delta_h1 = (W_h2.T @ delta_h2) * relu_derivative(Z_h1)

# gradien bobot H1
dW_h1 = (1 / n) * (delta_h1 @ X.T)
db_h1 = (1 / n) * np.sum(delta_h1, axis=1, keepdims=True)

print("=" * 50)
print("HASIL BACKWARD PASS")
print("=" * 50)
print(f"\ndelta_O per sampel : {delta_o}")
print(f"\ndL/dW_o  : {dW_o}")
print(f"dL/db_O  : {db_o.flatten()}")
print(f"\ndelta_H2 per sampel:\n{delta_h2}")
print(f"\ndL/dW_h2:\n{dW_h2}")
print(f"dL/db_H2 : {db_h2.flatten()}")
print(f"\ndelta_H1 per sampel:\n{delta_h1}")
print(f"\ndL/dW_h1:\n{dW_h1}")
print(f"dL/db_H1 : {db_h1.flatten()}")
```

### B.3 Weight Update


```python
W_h1_new = W_h1 - lr * dW_h1
W_h2_new = W_h2 - lr * dW_h2
W_o_new  = W_o  - lr * dW_o
b_h1_new = b_h1 - lr * db_h1
b_h2_new = b_h2 - lr * db_h2
b_o_new  = b_o  - lr * db_o

print("=" * 50)
print("UPDATE BOBOT (lr = 0.01)")
print("=" * 50)

print("\n-- Output Layer --")
print(f"W_o : {W_o.flatten()} -> {W_o_new.flatten()}")
print(f"b_O : {b_o.flatten()} -> {b_o_new.flatten()}")

print("\n-- Hidden Layer 2 --")
print(f"W_h2 sebelum:\n{W_h2}")
print(f"W_h2 sesudah:\n{W_h2_new}")
print(f"b_H2: {b_h2.flatten()} -> {b_h2_new.flatten()}")

print("\n-- Hidden Layer 1 --")
print(f"W_h1 sebelum:\n{W_h1}")
print(f"W_h1 sesudah:\n{W_h1_new}")
print(f"b_H1: {b_h1.flatten()} -> {b_h1_new.flatten()}")

print("\n" + "=" * 50)
print("Parameter yang benar-benar berubah:")
print("=" * 50)
ada_perubahan = False
for nama, lama, baru in [
    ("W_o",  W_o,  W_o_new),
    ("b_O",  b_o,  b_o_new),
    ("W_h2", W_h2, W_h2_new),
    ("b_H2", b_h2, b_h2_new),
    ("W_h1", W_h1, W_h1_new),
    ("b_H1", b_h1, b_h1_new),
]:
    if not np.allclose(lama, baru):
        ada_perubahan = True
        print(f"  {nama}: {lama.flatten()} -> {baru.flatten()}")

if not ada_perubahan:
    print("  (tidak ada perubahan)")
```

### B.4 Verifikasi Hasil Manual vs Python

Setiap nilai dihitung manual di bagian A dibandingkan satu per satu dengan output Python.


```python
import math

checks = [
    # forward sample 1
    ("Z_H1_1 sample1",   Z_h1[0,0], -10.0),
    ("A_H1_1 sample1",   A_h1[0,0],   0.0),
    ("Z_H1_2 sample1",   Z_h1[1,0],   0.0),
    ("A_H1_2 sample1",   A_h1[1,0],   0.0),
    ("Z_H1_3 sample1",   Z_h1[2,0],   6.0),
    ("A_H1_3 sample1",   A_h1[2,0],   6.0),
    ("Z_H2_1 sample1",   Z_h2[0,0], -11.0),
    ("A_H2_1 sample1",   A_h2[0,0],   0.0),
    ("Z_H2_2 sample1",   Z_h2[1,0], -11.0),
    ("A_H2_2 sample1",   A_h2[1,0],   0.0),
    ("y_hat sample1",    Y_hat[0,0],  1.0),
    # forward sample 2
    ("Z_H1_1 sample2",   Z_h1[0,1],   3.0),
    ("A_H1_1 sample2",   A_h1[0,1],   3.0),
    ("Z_H1_2 sample2",   Z_h1[1,1],   2.0),
    ("A_H1_2 sample2",   A_h1[1,1],   2.0),
    ("Z_H1_3 sample2",   Z_h1[2,1],  -1.0),
    ("A_H1_3 sample2",   A_h1[2,1],   0.0),
    ("Z_H2_1 sample2",   Z_h2[0,1],  10.0),
    ("A_H2_1 sample2",   A_h2[0,1],  10.0),
    ("Z_H2_2 sample2",   Z_h2[1,1],   8.0),
    ("A_H2_2 sample2",   A_h2[1,1],   8.0),
    ("y_hat sample2",    Y_hat[0,1],  5.0),
    # loss
    ("MSE loss",         mse,          2.0),
    # gradien
    ("delta_O sample1",  delta_o[0,0], 2.0),
    ("delta_O sample2",  delta_o[0,1], 0.0),
    ("dL/dW1_o",         dW_o[0,0],    0.0),
    ("dL/dW2_o",         dW_o[0,1],    0.0),
    ("dL/db_O",          db_o[0,0],    1.0),
    ("delta_H2_1 s1",    delta_h2[0,0], 0.0),
    ("delta_H2_2 s1",    delta_h2[1,0], 0.0),
    ("delta_H2_1 s2",    delta_h2[0,1], 0.0),
    ("delta_H2_2 s2",    delta_h2[1,1], 0.0),
    # update
    ("b_O setelah update", b_o_new[0,0], 0.99),
]

print("=" * 60)
print("VERIFIKASI: Manual vs Python")
print("=" * 60)

semua_lulus = True
for nama, hasil, ekspektasi in checks:
    lulus = math.isclose(hasil, ekspektasi, abs_tol=1e-9)
    if not lulus:
        semua_lulus = False
    status = "PASS" if lulus else "FAIL"
    print(f"  [{status}]  {nama:<26} | ekspektasi: {ekspektasi:>7}  | python: {hasil:>7}")

print()
if semua_lulus:
    print("Semua nilai cocok - perhitungan manual sudah benar.")
else:
    print("Ada nilai yang tidak cocok - cek ulang perhitungan manual.")
```

## Kesimpulan

Hasil dari 1 step training pada data 4B:

| Aspek | Hasil |
|-------|-------|
| MSE sebelum update | 2.0 |
| Prediksi sample 1 (I1=-4, I2=3) | y_hat = 1, target = -1, error = 4 |
| Prediksi sample 2 (I1=1, I2=0) | y_hat = 5, target = 5, error = 0 |
| Parameter yang berubah | hanya b_O: 1.0 -> 0.99 |
| Semua bobot lain | tidak berubah, gradien = 0 |

**Kenapa hampir semua gradien = 0?**

Ada dua alasan yang terjadi bersamaan:

1. Dying ReLU di Hidden Layer 2 pada sample 1
   - z_H2_1 dan z_H2_2 keduanya = -11 (negatif)
   - ReLU dari nilai negatif = 0, dan turunannya juga 0
   - Karena turunannya 0, gradien tidak bisa mengalir ke belakang (chain rule berhenti di sini)
   - Padahal sample 1 punya error yang cukup besar (error = 4)

2. Sample 2 tidak menghasilkan error sama sekali
   - y_hat = y = 5, jadi delta_O = 0
   - Kalau delta_O = 0, semua gradien yang berasal dari sample ini juga 0

Gabungan keduanya membuat satu-satunya update yang terjadi hanyalah pada bias output.
